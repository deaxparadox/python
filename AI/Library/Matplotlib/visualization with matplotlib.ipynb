**Continous Errors**

```python
from sklearn.gaussian_process import GaussianProcess

# define the model and draw some data 
model = lambda x: x * np.sin(x)
xdata = np.array([1,3,5,6,8])
ydata = model(xdata)

# computethe Gaussian process fit
gp = GaussianProcess(corr='cubic', theta0=1e-2, theaL=1e-4, thetaY=1E-1, random_start=100)
fp.git(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2*np.sqrt(MSE)      # 2*sigma ~ 95% confidence region
```

```python
# Visualize the result 
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit-dyfit, yfit+dyfit, color='gray', alpha=0.2)
plt.xlim(0, 10)
```

# Density and Contour Plots

```python
"""
- display three-dimensional data in two dimensiona using contours aor color-coded regions
- Theere are three Matplotlib functions that canb e helpfull for this task: 
    1. plt.contour
        - for contour plots
    2. plt.contoutf
        - for filled contour pots 
    3. plt.imshow
        - for showing images
"""
```

**Visualizing a Three-Dimensional Function**

```python
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
"""A contour plot can be created with the plt.contour function. 
- It takes three argument 
    1. a grid of x values
    2. a grid of y values
    3. a grid of z values
- The x and y values represent positions on the plot,
- and the z values, will be represented by the contour levels.
"""

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black')
```

```python
"""you can color-code the lines by specifying a colormap with the cmap argument.
"""
plt.contour(X, Y, Z, 20, cmap='RdGy')
```

```python
"""
- We can change this by switching to a filled contour plot using the 
plt.contourf() function, which uses largely the same syntax as plt.contour()

- plt.colorbar() command, which automatically creates an additional axis with 
labeled color information for the plot.
"""

plt.contourf(X, Y, Z, 20, cmap="RdGy")
plt.colorbar()
```

```python
"""use the plt.imshow() function, which interprets a two dimensional grid of data as an image."""

plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis()
```

```python
"""
- There are a few potential gotchas with imshow(), however:
    1. plt.imshow() doesn't accept an x and y grid, so you
    must manually specify the extent[xmin, xmax, ymin, ymax] of the 
    image on the plot
    2. plt.imshow() by default follows the standard image array 
    definition where the origin is in the upper left, not in the 
    lower left as in most contour plots. This must be cahnaged when
    showing gridded data.
    3. plt.imshow() will automatically adjust the axis aspect ratio
    to match the input data; you can change this by setting, for example,
    plt.axis(aspect='image') to make x and y units match.
"""
```

```python
"""
- Combine contour plots and image plots
"""

contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
```

*Kernel density estimation*

```python
"""A KERNEL DENSITY REPRESENTATION OF A DISTRIBUTION

- Another common method of evaluating densities in multiple dimensions is kernel-density-estimation (KDE)

- One extremely equick and simple KDE implementation exists in the scipy.stats package.
"""

from scipy.stats import gaussian_kde

# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a regular gride
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape), origin='lower', aspect='auto', extent=[-3.5, 3.5, -6, 6], cmap='Blues')
cb = plt.colorbar()
cb.set_label('density')
```

# Choosing Elements for the legend

```python
"""
- we can fine tune which elements and labels appear in the legend by 
using the object returned by plot commands.

- The plot.plot() command is able to create multiple lines at one, and returns 
a list of created line instances

- Pasing any of these to plt.legend() will tell it which to identify along with the labels 
we'd like to specify.
"""

y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
lines  = plt.plot(x,y)

# lines is a list of plt.Line2D instances
plt.legend(lines[:2], ['first', 'second'])

"""Customization of legend elements"""
```

```python
plt.plot(x, y[:, 0], label='first')
plt.plot(x, y[:, 1], label='second')
plt.plot(x, y[:, 2])
plt.legend(framealpha=1, frameon=True)

# alternating method for customizing legend elements
```

```python
"""NOTE
By default, the legend ignores all elements without a label attribute set.
"""
```

**Multiple legends**

```python
"""
We can work around this by creating a new legend artist from scratch, and then using the lower-level
ax.add_artist() method to manually add the second artist tothe plot.
"""

fig, ax = plt.subplots()

lines = []
styles=['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2), 
                   styles[i], color='black')
ax.axis('equal')

# specify the lines and labels of the first legend
ax.legend(lines[:2], ['line A', 'line B'],
         loc='upper right', frameon=False)

# Create the second legend and add the artist manually
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
            loc='lower right', frameon=False)
ax.add_artist(leg)


'''split plot legend'''
```
