```python
# plt.plot([1,2,3,4])
plt.plot([1,2,3,4], [1,4,9,16])
plt.ylabel('some numbers')
plt.show()
```

```python
import numpy as np
```

```python
"""ploting several lines with differnt format styles in one command using arrays."""

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
